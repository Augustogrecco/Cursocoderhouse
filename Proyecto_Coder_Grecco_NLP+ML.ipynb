# NLP Preprocesamiento

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')  # Para asegurar compatibilidad en lematización en español
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Configuración de stopwords en español y lematizador
stop_words = set(stopwords.words('spanish'))
lemmatizer = WordNetLemmatizer()

# Función de preprocesamiento
def preprocess_text(text):
    # 1. Convertimos a minúsculas
    text = text.lower()

    # 2. Eliminamos puntuación
    text = text.translate(str.maketrans("", "", string.punctuation))

    # 3. Tokenización
    tokens = word_tokenize(text)

    # 4. Remoción de stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # 5. Lematización
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Reconstruimos el texto preprocesado
    return " ".join(tokens)

# Aplicamos la función de preprocesamiento a la columna de texto
data['Hallazgo_preprocesado'] = data['Hallazgo'].apply(preprocess_text)

# Mostrar algunas filas para verificar el resultado
data[['Hallazgo', 'Hallazgo_preprocesado']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Configuramos el vectorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Aplicamos TF-IDF al texto preprocesado
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Hallazgo_preprocesado'])

# Convertimos el resultado a un DataFrame para visualizar
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Mostramos las primeras filas de la matriz TF-IDF
tfidf_df.head()
```

```python
from textblob import TextBlob

# Función para asignar sentimiento
def obtener_sentimiento(texto):
    # Calcular la polaridad del sentimiento
    polarity = TextBlob(texto).sentiment.polarity
    # Definir la categoría de sentimiento en base a la polaridad
    if polarity > 0:
        return "positivo"
    elif polarity < 0:
        return "negativo"
    else:
        return "neutral"

# Aplicar la función de sentimiento a cada hallazgo y crear la columna 'Sentimiento'
data['Sentimiento'] = data['Hallazgo_preprocesado'].apply(obtener_sentimiento)

# Mostrar algunos resultados para verificar
data[['Hallazgo', 'Sentimiento']].head()
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Aplicar SMOTE al conjunto de entrenamiento para Bag of Words
smote = SMOTE(random_state=42)
X_train_bow_resampled, y_train_bow_resampled = smote.fit_resample(X_train_bow, y_train)
print("Distribución de clases después de SMOTE (Bag of Words):", Counter(y_train_bow_resampled))

# Aplicar SMOTE al conjunto de entrenamiento para TF-IDF
X_train_tfidf_resampled, y_train_tfidf_resampled = smote.fit_resample(X_train_tfidf, y_train)
print("Distribución de clases después de SMOTE (TF-IDF):", Counter(y_train_tfidf_resampled))
```

Paso 1: Entrenamiento y Evaluación del Modelo Naive Bayes

Naive Bayes con Bag of Words

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Inicializar el modelo Naive Bayes
nb_bow = MultinomialNB()
# Entrenar el modelo
nb_bow.fit(X_train_bow_resampled, y_train_bow_resampled)
# Predicción en el conjunto de prueba
y_pred_bow = nb_bow.predict(X_test_bow)

# Evaluar el rendimiento
print("Naive Bayes - Bag of Words")
print(classification_report(y_test, y_pred_bow))
print("Accuracy:", accuracy_score(y_test, y_pred_bow))
```

Naive Bayes con TF-IDF

```python
# Inicializar el modelo Naive Bayes
nb_tfidf = MultinomialNB()
# Entrenar el modelo
nb_tfidf.fit(X_train_tfidf_resampled, y_train_tfidf_resampled)
# Predicción en el conjunto de prueba
y_pred_tfidf = nb_tfidf.predict(X_test_tfidf)

# Evaluar el rendimiento
print("\nNaive Bayes - TF-IDF")
print(classification_report(y_test, y_pred_tfidf))
print("Accuracy:", accuracy_score(y_test, y_pred_tfidf))
```

Paso 2: Entrenamiento y Evaluación del Modelo SVM

SVM con Bag of Words

```python
from sklearn.svm import SVC

# Inicializar el modelo SVM
svm_bow = SVC(kernel='linear', random_state=42)
# Entrenar el modelo
svm_bow.fit(X_train_bow_resampled, y_train_bow_resampled)
# Predicción en el conjunto de prueba
y_pred_bow = svm_bow.predict(X_test_bow)

# Evaluar el rendimiento
print("\nSVM - Bag of Words")
print(classification_report(y_test, y_pred_bow))
print("Accuracy:", accuracy_score(y_test, y_pred_bow))
```

SVM con TF-IDF

```python
# Inicializar el modelo SVM
svm_tfidf = SVC(kernel='linear', random_state=42)
# Entrenar el modelo
svm_tfidf.fit(X_train_tfidf_resampled, y_train_tfidf_resampled)
# Predicción en el conjunto de prueba
y_pred_tfidf = svm_tfidf.predict(X_test_tfidf)

# Evaluar el rendimiento
print("\nSVM - TF-IDF")
print(classification_report(y_test, y_pred_tfidf))
print("Accuracy:", accuracy_score(y_test, y_pred_tfidf))
```

Paso 3: Entrenamiento y Evaluación del Modelo de Regresión Logística

Regresión Logística con Bag of Words

```python
from sklearn.linear_model import LogisticRegression

# Inicializar el modelo de Regresión Logística
lr_bow = LogisticRegression(random_state=42, max_iter=1000)
# Entrenar el modelo
lr_bow.fit(X_train_bow_resampled, y_train_bow_resampled)
# Predicción en el conjunto de prueba
y_pred_bow = lr_bow.predict(X_test_bow)

# Evaluar el rendimiento
print("\nRegresión Logística - Bag of Words")
print(classification_report(y_test, y_pred_bow))
print("Accuracy:", accuracy_score(y_test, y_pred_bow))
```

Regresión Logística con TF-IDF

```python
# Inicializar el modelo de Regresión Logística
lr_tfidf = LogisticRegression(random_state=42, max_iter=1000)
# Entrenar el modelo
lr_tfidf.fit(X_train_tfidf_resampled, y_train_tfidf_resampled)
# Predicción en el conjunto de prueba
y_pred_tfidf = lr_tfidf.predict(X_test_tfidf)

# Evaluar el rendimiento
print("\nRegresión Logística - TF-IDF")
print(classification_report(y_test, y_pred_tfidf))
print("Accuracy:", accuracy_score(y_test, y_pred_tfidf))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Función para mostrar la matriz de confusión
def mostrar_matriz_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negativo", "Positivo"])
    disp.plot(cmap="Blues")
    plt.title(title)
    plt.show()
```

TF-IDF no funciona tan bien en este conjunto de datos con SVM y Regresión Logística, especialmente para la clase negativa. Esto puede deberse a que la representación de Bag of Words es más efectiva para capturar patrones básicos en este caso específico de análisis de sentimientos en auditoría.

Ahora, vamos a crear una función que tome una lista de nuevos hallazgos, los transforme con el vectorizador y use el modelo para predecir si cada hallazgo es positivo o negativo.

```python
def predecir_sentimiento(hallazgos):
    # Transformar los hallazgos usando el vectorizador cargado
    hallazgos_vectorizados = vectorizador.transform(hallazgos)

    # Hacer predicciones
    predicciones = modelo.predict(hallazgos_vectorizados)

    # Mostrar resultados
    for hallazgo, prediccion in zip(hallazgos, predicciones):
        print(f"Hallazgo: {hallazgo}\nPredicción: {'Positivo' if prediccion == 'positivo' else 'Negativo'}\n")
```
