```python
df = pd.DataFrame(data)

# Resumen estadístico del DataFrame
print("\nResumen estadístico del DataFrame:")
print(df.describe())

# Verificar los tipos de datos de cada columna
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

Analizando el data frame, vamos a trabajar un poco con columna por columna, para quitar o ajustar los datos faltantes, segun el siguiente detalle:
En el caso de "temp min y temp max", vamos a imputar los valores faltantes la media de la temperatura mínima o máxima respectivamente para ese mes del año y esa Cuidad.

```python
# Paso 1: Convertir la columna 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Paso 2: Extraer el mes del año y crear una nueva columna para almacenarlo
df['Month'] = df['Date'].dt.month

# Paso 3: Calcular la media de la temperatura mínima y máxima para cada ubicación y mes
monthly_means = df.groupby(['Location', 'Month'])[['MinTemp', 'MaxTemp']].mean()

# Paso 4: Imputar los valores faltantes de la temperatura mínima y máxima utilizando la media correspondiente a cada ubicación y mes
for index, row in df.iterrows():
    location = row['Location']
    month = row['Date'].month
    if pd.isna(row['MinTemp']):
        location_month_mean = monthly_means.loc[(location, month)]['MinTemp']
        df.at[index, 'MinTemp'] = location_month_mean
    if pd.isna(row['MaxTemp']):
        location_month_mean = monthly_means.loc[(location, month)]['MaxTemp']
        df.at[index, 'MaxTemp'] = location_month_mean
# Ahora df contendrá los valores faltantes de MinTemp y MaxTemp imputados con la media de cada ubicación y mes
```

```python
# Resumen estadístico del DataFrame
print("\nResumen estadístico del DataFrame:")
print(df.describe())

# Verificar los tipos de datos de cada columna
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
columns_to_drop = ['Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm','Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm']

# Eliminar las columnas del DataFrame ya que tienen muchos valores faltantes, y no serán relevantes en el análisis.
df = df.drop(columns=columns_to_drop)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
# Reemplazar los valores faltantes de Rainfall con cero
df['Rainfall'].fillna(0, inplace=True)
# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
columns_to_drop = ['WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','WindSpeed3pm', 'Humidity3pm', 'Temp9am', 'Temp3pm',]

# Eliminar las columnas del DataFrame
df = df.drop(columns=columns_to_drop)
# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
# Lista de columnas donde se eliminarán los valores faltantes
columns_to_check = ['RainToday', 'RainTomorrow']

# Eliminar filas con valores faltantes en las columnas especificadas
df = df.dropna(subset=columns_to_check)

# df ahora contendrá el DataFrame sin las filas que contienen valores faltantes en las columnas especificadas
# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
# Paso 1: Calcular el promedio de WindSpeed9am por mes y ciudad
monthly_city_wind_mean = df.groupby(['Location', df['Date'].dt.month])['WindSpeed9am'].transform('mean')

# Paso 2: Calcular el promedio de Humidity9am por mes y ciudad
monthly_city_humidity_mean = df.groupby(['Location', df['Date'].dt.month])['Humidity9am'].transform('mean')

# Paso 3: Reemplazar los valores faltantes con los promedios calculados
df['WindSpeed9am'].fillna(monthly_city_wind_mean, inplace=True)
df['Humidity9am'].fillna(monthly_city_humidity_mean, inplace=True)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualización usando Seaborn con colores por ubicación
plt.figure(figsize=(10, 6))
sns.boxplot(x='Location', y='MinTemp', data=df, palette='Set2')
plt.title('Distribución de Temperatura Mínima por Ubicación')
plt.xlabel('Ubicación')
plt.ylabel('Temperatura Mínima (°C)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Location', y='MaxTemp', data=df, palette='Set2')
plt.title('Distribución de Temperatura Máxima por Ubicación')
plt.xlabel('Ubicación')
plt.ylabel('Temperatura Máxima (°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' con las columnas 'Humidity9am' y 'RainToday'

# Filtrar datos para días lluviosos y días no lluviosos
humedad_dias_lluviosos = df[df['RainToday'] == 'Yes']['Humidity9am']
humedad_dias_no_lluviosos = df[df['RainToday'] == 'No']['Humidity9am']

# Crear gráfico de densidad para días lluviosos y días no lluviosos
plt.figure(figsize=(10, 6))
sns.kdeplot(humedad_dias_lluviosos, label='Días Lluviosos', shade=True)
sns.kdeplot(humedad_dias_no_lluviosos, label='Días No Lluviosos', shade=True)
plt.title('Distribución de Humedad a las 9 am')
plt.xlabel('Humedad a las 9 am')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' con las columnas 'Date', 'Rainfall' y 'Location'

# Extraer el mes y el año de la columna 'Date'
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Calcular el promedio anual para cada ciudad y mes
promedio_anual_por_ciudad_mes = df.groupby(['Location', 'Month'])['Rainfall'].mean().reset_index()

# Crear el gráfico de línea
plt.figure(figsize=(10, 6))
sns.lineplot(data=promedio_anual_por_ciudad_mes, x='Month', y='Rainfall', hue='Location', ci=None)
plt.title('Precipitación Promedio Mensual por Ciudad')
plt.xlabel('Mes')
plt.ylabel('Precipitación Promedio')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend(title='Ciudad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
