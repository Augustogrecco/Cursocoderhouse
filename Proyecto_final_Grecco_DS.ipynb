**2.	Análisis Exploratorio de Datos (EDA)**

•	Cargar los datos meteorológicos de las tres ciudades de Australia desde Github.


```python
import pandas as pd
import numpy as np

# URL del conjunto de datos en GitHub
url = 'https://raw.githubusercontent.com/Augustogrecco/Cursocoderhouse/main/Weather%20AUS%203%20CUIDADES.csv'

# Leer el conjunto de datos desde la URL
data = pd.read_csv(url, delimiter=',')

data
```

•	Descripción de las variables meteorológicas disponibles.

**Diccionario de variables:**

1. Date: La fecha en la que se registraron los datos.

2. Location: La ubicación geográfica donde se recopilaron los datos meteorológicos.

3. MinTemp: La temperatura mínima registrada durante el día en grados Celsius.

4. MaxTemp: La temperatura máxima registrada durante el día en grados Celsius.

5. Rainfall: La cantidad de lluvia registrada en milímetros.

6. Evaporation: La tasa de evaporación medida en milímetros por día.

7. Sunshine: El número de horas de sol registradas durante el día. 8. WindGustDir: La dirección de la ráfaga de viento más fuerte.

9. WindGustSpeed: La velocidad máxima del viento en kilómetros por hora. 10. WindDir9am: La dirección del viento a las 9:00 AM.

11. WindDir3pm: La dirección del viento a las 3:00 PM.

12. WindSpeed9am: La velocidad del viento a las 9:00 AM en kilómetros por hora.

13. WindSpeed3pm: La velocidad del viento a las 3:00 PM en kilómetros por hora.

14. Humidity9am: La humedad relativa a las 9:00 AM en porcentaje.

15. Humidity3pm: La humedad relativa a las 3:00 PM en porcentaje.

16. Pressure9am: La presión atmosférica a las 9:00 AM en hectopascales.

17. Pressure3pm: La presión atmosférica a las 3:00 PM en hectopascales.

18. Cloud9am: La cantidad de nubosidad a las 9:00 AM en octavos.

19. Cloud3pm: La cantidad de nubosidad a las 3:00 PM en octavos. 20. Temp9am: La temperatura a las 9:00 AM en grados Celsius.

21. Temp3pm: La temperatura a las 3:00 PM en grados Celsius.

22. RainToday: Variable binaria que indica si llovió ese día (Sí/No).

23. RainTomorrow: Variable binaria que indica si lloverá al día siguiente (Sí/No).

•	Exploración inicial de los datos.

```python
df = pd.DataFrame(data)

# Resumen estadístico del DataFrame
print("\nResumen estadístico del DataFrame:")
print(df.describe())

# Verificar los tipos de datos de cada columna
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

En base a este primer análisis, vamos a hacer algunos ajustes, para modelar los datos y poder continuar trabajando.

```python
# Paso 1: Convertir la columna 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Paso 2: Extraer el mes del año y crear una nueva columna para almacenarlo
df['Month'] = df['Date'].dt.month

# Paso 3: Calcular la media de la temperatura mínima y máxima para cada ubicación y mes
monthly_means = df.groupby(['Location', 'Month'])[['MinTemp', 'MaxTemp']].mean()

# Paso 4: Imputar los valores faltantes de la temperatura mínima y máxima utilizando la media correspondiente a cada ubicación y mes
for index, row in df.iterrows():
    location = row['Location']
    month = row['Date'].month
    if pd.isna(row['MinTemp']):
        location_month_mean = monthly_means.loc[(location, month)]['MinTemp']
        df.at[index, 'MinTemp'] = location_month_mean
    if pd.isna(row['MaxTemp']):
        location_month_mean = monthly_means.loc[(location, month)]['MaxTemp']
        df.at[index, 'MaxTemp'] = location_month_mean
# Ahora df contendrá los valores faltantes de MinTemp y MaxTemp imputados con la media de cada ubicación y mes

# Paso 5 Eliminar las columnas del DataFrame ya que tienen muchos valores faltantes, y no serán relevantes en el análisis.

columns_to_drop = ['Evaporation','WindSpeed9am', 'Humidity9am', 'Sunshine', 'Cloud9am', 'Cloud3pm','Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm','WindSpeed3pm', 'Humidity3pm', 'Temp9am', 'Temp3pm',]

df = df.drop(columns=columns_to_drop)

# Paso 6 COmo tenemos solo 10 días para este trabajo, se define usar la estrategia de reemplazar los valores faltantes de Rainfall con cero, aunque podríamos como recomendación a futuro, buscar por ejemplo la lluvia de lugares cercanos y armar un modelo de ML para competar estos valores faltantes con modelos de proximidad para esas fechas.

# Reemplazar los valores faltantes de Rainfall con cero
df['Rainfall'].fillna(0, inplace=True)

# Otra definición importante tomando en cuenta el tiempo de trabajo, es eliminar los registros donde no tenemos en 'RainToday', 'RainTomorrow', esperando que no tenga mayor impacto en el modelo. Si en un futuro se reconstruyen los valores faltantes estos campos igualmente se podrían predecir con modelos de ML.

# Lista de columnas donde se eliminarán los valores faltantes
columns_to_check = ['RainToday', 'RainTomorrow']

# Eliminar filas con valores faltantes en las columnas especificadas
df = df.dropna(subset=columns_to_check)
```

Revisamos nuevamente el data frame

```python
# Resumen estadístico del DataFrame
print("\nResumen estadístico del DataFrame:")
print(df.describe())

# Verificar los tipos de datos de cada columna
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Contar los valores perdidos en cada columna
print("\nValores perdidos en cada columna:")
print(df.isnull().sum())
```

**Resumen Estadístico del DataFrame modificado**

MinTemp (Temperatura Mínima): La temperatura mínima promedio registrada es de aproximadamente 12°C, con una variación de -3.3°C a 29°C.
MaxTemp (Temperatura Máxima): La temperatura máxima promedio registrada es de alrededor de 24.5°C, con una variación de 6.8°C a 47.3°C.
Rainfall (Lluvia): La cantidad promedio de lluvia es de aproximadamente 2.24 mm, con valores que van desde 0 mm hasta 240 mm.


No hay valores perdidos en ninguna de las columnas del DataFrame, lo que sugiere que los datos están completos y listos para su análisis.

•	Visualización de datos para comprender mejor las tendencias y patrones climáticos.


Histogramas de Temperaturas: Para visualizar la distribución de las temperaturas mínimas y máximas.

```python
import matplotlib.pyplot as plt

# Crear subplots para cada ciudad
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Histograma de Temperatura Mínima
for i, city in enumerate(['Newcastle', 'Moree', 'Albury']):
    axes[i, 0].hist(data[data['Location'] == city]['MinTemp'], bins=20, color='skyblue', edgecolor='black')
    axes[i, 0].set_title(f'Distribución de Temperatura Mínima - {city}')
    axes[i, 0].set_xlabel('Temperatura (°C)')
    axes[i, 0].set_ylabel('Frecuencia')
    axes[i, 0].grid(True)

# Histograma de Temperatura Máxima
for i, city in enumerate(['Newcastle', 'Moree', 'Albury']):
    axes[i, 1].hist(data[data['Location'] == city]['MaxTemp'], bins=20, color='salmon', edgecolor='black')
    axes[i, 1].set_title(f'Distribución de Temperatura Máxima - {city}')
    axes[i, 1].set_xlabel('Temperatura (°C)')
    axes[i, 1].set_ylabel('Frecuencia')
    axes[i, 1].grid(True)

plt.tight_layout()
plt.show()
```

Gráfico de Línea de Lluvia a lo largo del Tiempo para Cada Ciudad:


```python
# Crear subplots para cada ciudad
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

# Gráfico de Línea de Lluvia a lo largo del Tiempo
for i, city in enumerate(['Newcastle', 'Moree', 'Albury']):
    axes[i].plot(data[data['Location'] == city]['Date'], data[data['Location'] == city]['Rainfall'], color='green')
    axes[i].set_title(f'Cantidad de Lluvia a lo largo del Tiempo - {city}')
    axes[i].set_xlabel('Fecha')
    axes[i].set_ylabel('Lluvia (mm)')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

Diagrama de Cajas y Bigotes de Temperaturas para Cada Ciudad:

```python
import seaborn as sns

# Diagrama de Cajas y Bigotes de Temperatura Mínima y Máxima para cada ciudad
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

for i, city in enumerate(['Newcastle', 'Moree', 'Albury']):
    sns.boxplot(x='Location', y='MinTemp', data=data[data['Location'] == city], ax=axes[0])
    axes[0].set_title('Diagrama de Cajas y Bigotes de Temperatura Mínima')
    axes[0].set_ylabel('Temperatura Mínima (°C)')
    axes[0].set_xlabel('Ciudad')
    axes[0].grid(True)

    sns.boxplot(x='Location', y='MaxTemp', data=data[data['Location'] == city], ax=axes[1])
    axes[1].set_title('Diagrama de Cajas y Bigotes de Temperatura Máxima')
    axes[1].set_ylabel('Temperatura Máxima (°C)')
    axes[1].set_xlabel('Ciudad')
    axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Gráfico de Barras de Lluvia Hoy y Mañana para Cada Ciudad:

```python
# Gráfico de Barras de Lluvia Hoy y Mañana para cada ciudad
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

for i, city in enumerate(['Newcastle', 'Moree', 'Albury']):
    sns.countplot(x='RainToday', hue='RainTomorrow', data=data[data['Location'] == city], ax=axes[i], palette='coolwarm')
    axes[i].set_title(f'Lluvia Hoy vs. Lluvia Mañana - {city}')
    axes[i].set_xlabel('Lluvia Hoy')
    axes[i].set_ylabel('Frecuencia')
    axes[i].legend(title='Lluvia Mañana', loc='upper right')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Supongamos que tienes un DataFrame llamado 'df' con las columnas 'Date', 'Rainfall' y 'Location'

# Extraer el mes y el año de la columna 'Date'
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Calcular el promedio anual para cada ciudad y mes
promedio_anual_por_ciudad_mes = df.groupby(['Location', 'Month'])['Rainfall'].mean().reset_index()

# Crear el gráfico de línea
plt.figure(figsize=(10, 6))
sns.lineplot(data=promedio_anual_por_ciudad_mes, x='Month', y='Rainfall', hue='Location', ci=None)
plt.title('Precipitación Promedio Mensual por Ciudad')
plt.xlabel('Mes')
plt.ylabel('Precipitación Promedio')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend(title='Ciudad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
